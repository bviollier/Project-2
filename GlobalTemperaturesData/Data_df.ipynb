```python
import pandas as pd

country_path = "GlobalLandTemperaturesByCountry.csv"
major_city_path = "GlobalLandTemperaturesByMajorCity.csv"
state_path = "GlobalLandTemperaturesByState.csv"
global_path = "GlobalTemperatures.csv"

country_df = pd.read_csv(country_path)
major_city_df = pd.read_csv(major_city_path)
state_df = pd.read_csv(state_path)
global_df = pd.read_csv(global_path)
```

```python
#country_df
zimb_country_df = country_df.loc[country_df['Country']=='Zimbabwe']
zimb_country_df.head(10)

country_df = country_df[['dt','AverageTemperature','Country']]
country_df = country_df.rename(columns={
    "dt":"date",
    "AverageTemperature":"avgtempcountry",
    "Country":"country"
})

country_df.head()
```

```python
#major_city_df
paris_city_df = major_city_df.loc[major_city_df['City']=='Paris']
paris_city_df

major_city_df = major_city_df[['dt','AverageTemperature','City','Country']]
major_city_df = major_city_df.rename(columns={
    'dt':'date',
    'AverageTemperature':'avgtempcity',
    'City':'city',
    'Country':'country'
})
major_city_df.head()
```

```python
#state_df
cali_city_df = state_df.loc[state_df['State']=='California']
cali_city_df

state_df = state_df[['dt','AverageTemperature','State','Country']]
state_df = state_df.rename(columns={
    'dt':'date',
    'AverageTemperature':'avetempstate',
    'State':'state',
    'Country':'country'
})
state_df.head()
```

```python
#global_df
global_df = global_df.rename(columns={
    "dt":"date",
    "LandAverageTemperature":"avglandtemp",
    "LandAverageTemperatureUncertainty":" ",
    "LandMaxTemperature":"do i care",
    "LandMaxTemperatureUncertainty":" ",
    "LandMinTemperature":"no",
    "LandMinTemperatureUncertainty":" ",
    "LandAndOceanAverageTemperature":"avglandoceantemp",
    "LandAndOceanAverageTemperatureUncertainty":" "
})
```

```python
global_df = global_df[['date','avglandtempglobal','avglandoceantempglobal']]
```
